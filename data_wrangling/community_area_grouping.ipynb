```python
import pandas as pd 

df = pd.read_csv('traffic_analysis.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# drop unnecessary columns
df = df.drop(columns=["raw_row_number", "time", "location", "lat", "lng", "geocode_source", "beat", "district", "subject_age", "subject_race", "subject_sex", "officer_race", "officer_sex", "officer_years_of_service", "type", "violation", "outcome", "search_conducted", "search_person", "search_vehicle", "raw_race", "raw_driver_race", "geometry"])
# Display the first few rows of the modified dataframe
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# make a year column
df["date"] = pd.to_datetime(df["date"])
df["year"] = df["date"].dt.year
df.head()
```

```python
df.tail()
```

```python
# make citation numeric and not T/F
df["citation_numeric"] = df["citation_issued"].astype(int)
```

```python
# group by community area only
community_area_df = (
    df.groupby(["community_area_id"], as_index=False)
      .agg(
          citation_pct=("citation_numeric", "mean"),
          n_stops=("citation_numeric", "size"),
          community_area_name=("community_area_name", "first"),
          drug_abuse=("drug_abuse", "first"),
          major_crime=("major_crime", "first"),
          public_crime=("public_crime", "first"),
          violent_crime=("violent_crime", "first"),
          perceived_neighborhood_violence_rate=("perceived_neighborhood_violence_rate", "first"),
          perceived_neighborhood_safety_rate=("perceived_neighborhood_safety_rate", "first"),
          trust_in_law_enforcement=("trust_in_law_enforcement", "first"),
          trust_in_law_enforcement_rate=("trust_in_law_enforcement_rate", "first"),
          trust_in_local_government=("trust_in_local_government", "first"),
          trust_in_local_government_rate=("trust_in_local_government_rate", "first"),
          economic_diversity_index=("economic_diversity_index", "first"),
          hardship_index=("hardship_index", "first"),
          median_household_income=("median_household_income", "first"),
          limited_english_proficiency=("limited_english_proficiency", "first"),
          population=("population", "first"),
          poverty_rate=("poverty_rate", "first"),
          foreign_born=("foreign_born", "first"),
          demographics=("demographics", "first")
      )
)

community_area_df.head()
```
